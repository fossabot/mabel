```python
from mabel.data import Reader, STORAGE_CLASS, DictSet, SqlReader
from mabel.adapters.disk import DiskReader
```

```python
SQL = "SELECT apples FROM (SELECT COUNT(*) AS apples FROM data/complex GROUP BY tweet.location)"
SQL = "SELECT MAX(tweet.sentiment) as max_sen FROM data/complex GROUP BY tweet.location"
SQL = "SELECT count(*), user_name from data/tweets GROUP BY user_name HAVING COUNT(*) > 1"
SQL = "SELECT count(*) FROM  data/tweets GROUP BY user_verified"
SQL = "SELECT count(*) FROM tests.data.index.is GROUP BY user_verified"

tweets = SqlReader(
    SQL,
    inner_reader=DiskReader,
    raw_path=True,
    persistence=STORAGE_CLASS.MEMORY)

tweets
```
